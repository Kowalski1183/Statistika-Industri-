```python

grouped_df = df.groupby("Region").mean()
grouped_df.reset_index(inplace=True)  # Resetting index to include 'Region' as a column
counts = df.groupby("Region").size()
counts = counts.reset_index(name='Count')

# Merge the counts with the grouped dataframe on 'Region'
grouped_df = pd.merge(grouped_df, counts, on='Region')

print(grouped_df)
```
